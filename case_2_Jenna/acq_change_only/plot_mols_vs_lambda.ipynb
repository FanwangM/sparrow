```python
import matplotlib.pyplot as plt

lambdas = [1,3,5,12,13,14,15,20,30,40]

num_mols = [0,0,2,3,4,4,4,5,5,5]

num_mats = [0,0,4,7,12,12,12,18,18,18]

color = 'tab:blue'
fig, ax1 = plt.subplots(figsize=(4, 4))
ax1.set_xlabel("$\lambda_{1}$")
ax1.set_ylabel('Number of molecules selected', color = color)
ax1.plot(lambdas, num_mols, '-o', color = color)
ax1.tick_params(axis ='y', labelcolor = color)
ax1.arrow(7, 2.5, -3, 0, color=color, length_includes_head=True, head_width = 0.15, head_length = 0.8)

# Adding Twin Axes to plot using dataset_2
ax2 = ax1.twinx()
 
color = 'tab:red'
ax2.set_ylabel('Number of chemicals used', color = color)
ax2.plot(lambdas, num_mats, '--o', color = color)
ax2.tick_params(axis ='y', labelcolor = color)
ax1.arrow(16.5, 3.5, 3, 0, color=color, length_includes_head=True, head_width = 0.15, head_length = 0.8)

ax1.set_title("Effect of $\lambda_1$ on selected routes")
ax1.text(25, 0, "$\lambda_2=\lambda_3=\lambda_4=1$")

plt.savefig("case_2.png", bbox_inches='tight', dpi=1000)
plt.show()
```
